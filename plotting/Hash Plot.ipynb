```python
import numpy as np
import numpy.linalg as npla
import matplotlib.pyplot as plt

data = np.loadtxt("../data/hashdata.txt", dtype=int)
x = data[:, 0].T
y1 = data[:, 1].T
y2 = data[:, 2].T

plt.figure(figsize=(15, 10))
plt.scatter(x, y2, label="Prime Mod", color="blue")
plt.xlabel("Value")
plt.ylabel("Hash Value")
plt.title("Prime Mod Hash Distribution")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(15, 10))
plt.scatter(x, y1, label="Mersenne Prime", color="red")
plt.xlabel("Value")
plt.ylabel("Hash Value")
plt.title("Mersenne Prime Hash Distribution")
plt.legend()
plt.grid()
plt.show()
```
